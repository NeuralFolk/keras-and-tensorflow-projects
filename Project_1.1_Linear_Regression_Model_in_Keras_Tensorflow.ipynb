```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

X = np.arange(1, 50, 5)
```

```python
y = X + 10
```

```python
plt.scatter(X, y)
```

```python
linear_regression_model = tf.keras.Sequential()
```

```python
linear_regression_model.add(tf.keras.layers.Dense(10, input_shape=[1], name='input_layer'))
linear_regression_model.add(tf.keras.layers.Dense(1, name='output_layer'))
```

```python
linear_regression_model.compile(loss='mae', optimizer='sgd', metrics=['mae'])
```

```python
linear_regression_model.fit(X, y, epochs=10)
```

```python
linear_regression_model.predict([9])
```

```python
linear_regression_model = tf.keras.Sequential()

linear_regression_model.add(tf.keras.layers.Dense(20, input_shape=[1], activation='relu', name='input_layer'))
linear_regression_model.add(tf.keras.layers.Dense(1, activation='linear', name='output_layer'))

adam = tf.keras.optimizers.Adam(learning_rate=0.01)

linear_regression_model.compile(loss='mae', optimizer=adam, metrics=['mae'])
```

```python
linear_regression_model.fit(X, y, epochs=300)
```

```python
linear_regression_model.predict([23])
```

```python
X_test = X[80:]
y_test = y[80:]
```

```python
lrmld = tf.keras.Sequential(name="NeuralFolk")

lrmld.add(tf.keras.layers.Dense(100, input_shape=[1], name='input_layer', activation='relu'))
lrmld.add(tf.keras.layers.Dense(50, name='first_hidden_layer', activation='relu'))
lrmld.add(tf.keras.layers.Dense(25, name='sedond_hidden_layer', activation='relu'))
lrmld.add(tf.keras.layers.Dense(1, name='output_layer'))

adam = tf.keras.optimizers.Adam(learning_rate=0.001)

lrmld.compile(loss='mae', optimizer=adam, metrics=['mae'])
```

```python
lrmld.summary()
```

```python
history = lrmld.fit(X, y, epochs=500, validation_data = (X_test, y_test), verbose = 1)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
```

```python
plt.plot(history.history['mae'])
```

```python
tf.keras.utils.plot_model(lrmld, show_shapes = True)
```
